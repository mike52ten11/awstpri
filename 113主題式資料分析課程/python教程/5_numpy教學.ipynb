```python
import numpy as np
```

```python
a = np.array([
    [1,2,3],
    [4,5,6]
], dtype=np.float32)
```

```python
print(a.ndim)#數據維度
```

```python
print(a.shape)#數據形狀(列x行)
```

```python
print(a.dtype)#數據型態
```

```python
print(np.average(a))#平均
```

```python
a.reshape(-1)#變成1維
```

```python
a.reshape((3,2))#變成3x2
```

```python
a.T#a轉置
```

```python
a.transpose((1,0))#a 列轉置
```

```python
a.transpose((0,1))#a 行轉置
```
